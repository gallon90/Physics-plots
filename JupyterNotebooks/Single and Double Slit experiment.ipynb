```python
%matplotlib inline

from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-3,3,10000)

B=input('enter individual slit width in micrometers: ')
b=int(B)*10**(-4)

D=input('enter separation between both slits in millimeters: ')
d=int(D)*10**(-3)

Lambduh=input('enter wavelength in nanometers: ')
lambduh=int(Lambduh)**10**(-9)

L=.0003
I_o=1
#(37110)

#sin(theta) portion of slit equation is approximately equal to any x divided by L distance to detector(in meters)

alpha=(np.pi*d/lambduh)*x1/L

beta=(np.pi*b/lambduh)*x1/L

t=(np.cos(alpha))**2

prod = np.sinc(beta)**2

plt.figure()
plt.title('Single-slit Distribution')
plt.xlabel('$L\lambda/b$')
plt.ylabel('Arbitrary Intensity $I_0$ (Lux)')
plt.plot(x1, I_o*prod,"r")
#plt.savefig('singleslit.png', dpi=1000)


I=I_o*t*prod
plt.figure()
plt.title('Double-slit Distribution')
plt.xlabel('$L\lambda/b$')
plt.ylabel('Arbitrary Intensity $I_0$ (Lux)')
plt.plot(x1,I,'g')
#plt.savefig('doubleslit.png', dpi=1000)


```

```python
##The following shows that the distribution for 2 slits is a product of the 1st slit with cos(theta)**2
plt.figure()
plt.title('Superimposed Distribution of Single and Double Slits')
plt.xlabel(' $L\lambda/b$')
plt.ylabel('Arbitrary Intensity $I_0$ (Lux)')
plt.plot(x1,I,'g')
plt.plot(x1,I_o*prod,'--r')

#plt.savefig('superimposed.png', dpi=1000)
```

```python
import pylab as p
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.mplot3d.axes3d as p3
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)

wavelength = 4.0e-3
scrnDist =4e-2
scrnWdth = 2.4
srceSepn= 1.3e-2

xs=[0,0]
ys=[-srceSepn/2,srceSepn/2]
A=10
     
N=1000

dX=scrnDist*10/N
Xcoords=np.arange(dX,scrnDist+2*dX,dX)
dY=scrnDist*10/N
Ycoords=np.arange(-scrnDist/2,scrnDist/2+2*dY,dY)
[xd,yd]=np.meshgrid(Xcoords,Ycoords)
r1=np.sqrt((xd-xs[1])**2+(yd-ys[1])**2)
r2=np.sqrt((xd-xs[1])**2+(yd-ys[1])**2)
E0=A*np.sin(2*np.pi*r1/wavelength)/r1+A*np.sin(2*np.pi*r2/wavelength)/r2

fig=p.figure()
Efield=p3.Axes3D(fig)
Efield.plot_surface(xd,yd,E0,color='#2980B9')
Efield.view_init(elev=40, azim=200)
Efield.set_yticklabels([])
Efield.set_xticklabels([])
Efield.set_zticklabels([])

plt.show()
#plt.savefig('2wave.png', dpi=1000)
```
