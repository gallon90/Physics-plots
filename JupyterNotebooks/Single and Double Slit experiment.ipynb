```python
%matplotlib inline
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-3,3,10000)

B=input('enter individual slit width in micrometers: ')
b=int(B)*10**(-4)

D=input('enter separation between both slits in millimeters: ')
d=int(D)*10**(-3)

Lambduh=input('enter wavelength in nanometers: ')
lambduh=int(Lambduh)**10**(-9)

L=.0003
I_o=(37110)

#sin(theta) portion of slit equation is approximately equal to any x divided by L distance to detector(in meters)

alpha=(np.pi*d/lambduh)*x1/L

beta=(np.pi*b/lambduh)*x1/L

t=(np.cos(alpha))**2

prod = np.sinc(beta)**2



#if you want to save your figure, uncomment the plt.savefig portions

plt.figure()
plt.title('Single-slit Distribution')
plt.xlabel('Dispersion/(distance to detector)')
plt.ylabel('Intensity(Lux)')
plt.plot(x1, I_o*prod,"r")
#plt.savefig('singleslit.png', dpi=1000)


I=I_o*t*prod
plt.figure()
plt.title('Double-slit Distribution')
plt.xlabel('Dispersion/(distance to detector)')
plt.ylabel('Intensity(Lux)')
plt.plot(x1,I,'g')
#plt.savefig('doubleslit.png', dpi=1000)


```

```python

#The following shows that the distribution for 2 slits is a product of the 1st slit with cos(theta)**2
plt.figure()
plt.title('Superimposed Distribution')
plt.xlabel('Dispersion/(distance to detector)')
plt.ylabel('Intensity(Lux)')
plt.plot(x1,I,'g')
plt.plot(x1,I_o*prod,'--r')
#plt.savefig('superimposed.png', dpi=1000)
```
